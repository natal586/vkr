Косинусное сходство

```python
#similarity_matrix = linear_kernel(tested_companies_matr,preferred_markets_matr)
```

```python
cosine_matrix=[]
for i in range(0,15):
  company_dist = []
  for y in range(0, len(preferred_markets)):
    vec1=np.array(tested_companies_matr.iloc[i]).reshape(1,-1)
    vec2=np.array(preferred_markets_matr.iloc[y]).reshape(1,-1)
    dist = float(cosine_similarity(vec1,vec2))
    company_dist.append(dist)
  cosine_matrix.append(company_dist)
```

```python
cosine_matrix
```

```python
mapping = pd.Series(tested_companies_matr.index,index = tested_companies['Comp_name'])
mapping
```

```python
def recommend_based_on_cosine_similarity(comp_input):
  comp_index = mapping[comp_input]
  similarity_score = list(enumerate(cosine_matrix[comp_index]))
  similarity_score = sorted(similarity_score, key=lambda x: x[1], reverse=True)
  similarity_score = similarity_score[0:5]
  comp_indices = [i[0] for i in similarity_score]
  return (preferred_markets['Name'].iloc[comp_indices])
```
